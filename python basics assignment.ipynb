Q9) Write a Python program to convert user input from string to integer, float, and boolean types.

```python
a = str(input("Give a number:"))
print("The number given is:",a)
print("The type of the number is:",type(a))
b= int(a)
print("Converting str to int: ",type(b))
c = float(a)
print("Converting int to float: ",type(c))
d = bool(a)
print("Converting str to boolean types: ",type(d))
print(d)
```
